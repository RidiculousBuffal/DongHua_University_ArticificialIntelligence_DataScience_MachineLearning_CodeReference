## pandas快速绘图

  - `'barh'`：水平条形图。
  - `'hist'`：直方图。
  - `'box'`：箱线图。
  - `'scatter'`：散点图。
  - `'area'`：面积图。
  - `'pie'`：饼图。
  - `'kde'`：核密度估计图。
  - 其他类型也可以根据需要使用。

- `x`：指定用作x轴数据的列名或索引。可以是字符串或可迭代对象。

- `y`：指定用作y轴数据的列名或索引。可以是字符串或可迭代对象。

- `title`：图形的标题。

- `xlabel`：x轴的标签。

- `ylabel`：y轴的标签。

- `color`：指定图形的颜色。可以是字符串（如`'red'`、`'blue'`）或颜色码（如`'#FF0000'`）。

- `colormap`：指定用于绘制颜色映射的色彩图。

- `fontsize`：指定标题和标签的字体大小。

- `legend`：控制是否显示图例（默认为`True`）。

- `grid`：控制是否显示网格线（默认为`False`）。

- `figsize`：指定图形的大小，以元组`(width, height)`的形式表示。

- `rot`：旋转x轴标签的角度。

这些只是`df.plot()`函数的一些常用参数，还有其他参数可用于进一步自定义图形的外观和行为。你可以根据具体的需求选择和调整这些参数来生成符合要求的图形。

### 使用子图
>`add_subplot()`是Matplotlib中Figure对象的方法之一，用于在一个Figure对象中添加子图（subplot）。它接受多个参数来控制子图的位置、布局和其他属性。下面是`add_subplot()`方法的一些常用参数：

- `nrows`：指定子图的行数。

- `ncols`：指定子图的列数。

- `index`：指定当前子图的索引位置。索引从左上角开始，从左到右，从上到下递增。

- `rowspan`：指定当前子图所占的行数。

- `colspan`：指定当前子图所占的列数。

- `sharex`：指定是否共享x轴刻度。可以是一个坐标轴对象或布尔值。

- `sharey`：指定是否共享y轴刻度。可以是一个坐标轴对象或布尔值。

- `projection`：指定子图的投影方式，例如 `'3d'` 表示三维子图。

- `**kwargs`：其他关键字参数，用于进一步自定义子图的属性，如标题、标签、轴范围等。

通过使用这些参数，你可以创建自定义布局的Figure对象，并在其中添加不同位置和大小的子图。这样可以灵活地组织多个图形，并在一个图像中展示多个数据视图或子图。

````python
data=Series([41.3,48.9,54.0,59.5,64.4,68.9,74.4,82.1,90.0,
98.7,101.6], index=['2010','2011','2012','2013','2014',
'2015','2016','2017','2018','2019','2020'])
fig = plt.figure(figsize=(6,6))
# 绘制子图1
ax1=fig.add_subplot(2,1,1)
ax1.plot(data)
#绘制子图2
ax2 = fig.add_subplot(2,2,3)
# 使用Pandas绘制柱状图
data.plot(kind='bar',use_index=True,fontsize='small',ax=ax2)
ax3=fig.add_subplot(2,2,4)
#用pandas绘制箱型图
data.plot(kind='box',fontsize='small',xticks=[],ax=ax3)
plt.savefig('./out/output4.png',dpi=600)
````

> `annotate()`函数是Matplotlib库中用于添加注释文本的函数。它可以在图形中的指定位置添加注释，例如标记数据点或提供额外的说明。`annotate()`函数接受多个参数来控制注释的位置、外观和其他属性。下面是`annotate()`函数的一些常用参数：
- `s`：指定注释文本的内容。
- `xy`：指定注释的目标位置，以元组`(x, y)`的形式表示。
- `xytext`：指定注释文本的位置，以元组`(x, y)`的形式表示。默认为与`xy`相同。
`：指定`xy`和`xytext`的坐标系类型。常见选项包括：
  - `'data'`：数据坐标系（默认）。
  - `'axes fraction'`：轴的分数坐标系。
  - `'figure fraction'`：图形的分数坐标系。
- `textcoords`：指定注释文本的坐标系类型。与`xycoords`类似，常见选项也包括`'data'`、`'axes fraction'`和`'figure fraction'`。
- `arrowprops`：指定注释箭头的属性，如颜色、宽度、样式等。可以是一个字典或关键字参数。
- `fontsize`：指定注释文本的字体大小。
- `color`：指定注释文本的颜色。
- `ha`：指定注释文本的水平对齐方式。常见选项包括：
  - `'center'`：居中对齐（默认）。
  - `'left'`：左对齐。
  - `'right'`：右对齐。
- `va`：指定注释文本的垂直对齐方式。常见选项包括：
  - `'center'`：居中对齐（默认）。
  - `'top'``：顶部对齐。
  - `'bottom'`：底部对齐。
这些只是`annotate()`函数的一些常用参数，还有其他参数可用于进一步自定义注释的外观和行为。通过使用这些参数，你可以在图形中添加注释，提供额外的信息或突出显示特定的数据点。

### 散点图
`DataFrame.plot()` 是 Pandas 库中 DataFrame 对象的一个方法，用于绘制数据的图表。它提供了多种参数用于控制图表的样式、类型和显示方式。下面是一些常用的参数说明：

- `kind`：指定绘图类型的参数，可以是字符串或者一个图表类型常量。常用的图表类型包括：
  - `'line'`：线图（默认）
  - `'bar'`：柱状图
  - `'barh'`：水平柱状图
  - `'hist'`：直方图
  - `'box'`：箱线图
  - `'scatter'`：散点图
  - `'pie'`：饼图

- `x` 和 `y`：指定绘图时所使用的数据列。`x` 参数是可选的，当绘制柱状图、线图等需要指定横轴的图表时，可以使用该参数指定横轴的数据列。

- `subplots`：布尔值，表示是否为每个列绘制单独的子图，默认为 False。如果设置为 True，每个列将在独立的子图中绘制。

- `title`：图表的标题。

- `xlabel` 和 `ylabel`：横轴和纵轴的标签。

- `xlim` 和 `ylim`：指定横轴和纵轴的显示范围。

- `grid`：布尔值，表示是否显示网格线，默认为 False。

- `legend`：布尔值，表示是否显示图例，默认为 True。

- `color`：用于指定图表颜色的参数，可以是字符串表示的颜色名称，也可以是一个颜色列表。

- `figsize`：用于指定图表的大小，以元组 (width, height) 的形式指定。

- `rot`：旋转 x 轴刻度标签的角度。

`plt.scatter()` 是 Matplotlib 库中用于绘制散点图的函数。它提供了多种参数用于控制散点图的样式和显示方式。下面是一些常用的参数说明：

- `x` 和 `y`：指定散点图的 x 轴和 y 轴数据。

- `s`：散点的大小。可以是一个标量，表示所有散点的大小相同；也可以是一个数组，表示每个散点的大小不同。

- `c`：散点的颜色。可以是一个标量或一个数组，表示每个散点的颜色不同。常见的取值包括：
  - 单个颜色字符串，如 `'red'`、`'blue'`，表示所有散点的颜色相同。
  - 单个颜色字符，如 `'r'`、`'b'`，表示所有散点的颜色相同，颜色缩写。
  - 数组，可以是数字数组或字符串数组，表示每个散点的颜色不同。

- `marker`：散点的标记样式。常见的取值包括：
  - `'o'`：圆形（默认）
  - `'s'`：正方形
  - `'+'`：加号
  - `'x'`：叉号
  - `'d'`：菱形

- `alpha`：散点的透明度，取值范围为 0（完全透明）到 1（完全不透明）之间。

- `linewidths`：散点的边界线宽度。

- `edgecolors`：散点的边界颜色。

- `cmap`：用于指定颜色映射的参数。可以是一个字符串，表示使用预定义的颜色映射；也可以是一个 Colormap 对象，用于自定义颜色映射。

- `label`：散点图的标签，用于图例显示。

- `alpha`：散点的透明度，取值范围为 0（完全透明）到 1（完全不透明）之间。

- `vmin` 和 `vmax`：用于设置颜色映射的值范围。


````python
#从student.csv文件中读入数据绘制散点图
stData = pd.read_csv('./data/students.csv')
stData.plot(kind='scatter', x='Height', y='Weight', color='red', label='(Height,Weight)',
            title='Students Body Shape',marker = '*',grid=True,xlim=[150,200],ylim=[40,80])
plt.savefig('./out/output6.png',dpi=600)
````

### 柱状图

````python
# 采用柱状图比较不同性别在各年度出生的人口数量以及平均值
data = pd.read_csv('./data/Population.csv',index_col='Year')
birth = data[['Boys','Girls']]
mean = np.mean(birth,axis=0)
std = np.std(birth,axis=0)
# 创建图
fig = plt.figure(figsize=(6,2))
plt.subplots_adjust(wspace=0.6)
# 绘制均值的垂直和水平柱状图,标准差用误差线来表示
ax1 = fig.add_subplot(1,2,1)
mean.plot(kind='bar',color='cadetblue',yerr=std,ax=ax1,alpha=0.7,ecolor='k',title='Average of Births',rot=45)
ax2 = fig.add_subplot(1,2,2)
mean.plot(kind='barh',xerr=std,color='cadetblue',title='Average of Births',ax=ax2,alpha=0.7,ecolor='k',rot=45)
# 绘制复式柱状图和堆叠柱状图
birth.plot(kind='bar',title='Births of Boys & Girls')
birth.plot(kind='bar',stacked=True,title='Births of Boys & Girls')
plt.savefig('./out/output9.png',dpi=600,bbox_inches='tight')
````

### 饼图
在 Pandas 中，绘制饼图时可以使用 `pd.DataFrame.plot.pie()` 方法。以下是一些常用的参数说明：
- `y`：指定要绘制饼图的数据，可以是一个 Series 或 DataFrame。
- `labels`：标签数组或索引，用于指定每个扇区的标签。如果不提供标签，将使用索引标签。
- `autopct`：用于显示每个扇区占比的格式化字符串，例如 "%.1f%%"。默认值为 None，表示不显示占比。
- `startangle`：饼图的起始角度，以度数表示。默认为 0，表示从水平方向开始绘制。
- `counterclock`：布尔值，表示是否按逆时针方向绘制饼图。默认为 True。
- `explode`：用于突出显示某个扇区的偏移量数组。默认值为 None，表示所有扇区均不突出。
- `colors`：指定饼图扇区的颜色。可以是颜色字符串列表或 colormap 对象。
- `shadow`：布尔值，表示是否在饼图下方绘制阴影。默认为 False。
- `radius`：饼图的半径。默认为 1。
- `pctdistance`：占比标签与圆心的距离的比例。默认为 0.6。
- `labeldistance`：标签与圆心的距离的比例。默认为 1.1。
- `frame`：布尔值，表示是否绘制饼图的轴线和标签。默认为 None，表示自动确定是否绘制。
- `figsize`：用于指定图表的大小，以元组 (width, height) 的形式指定。
- `legend`：布尔值，表示是否显示图例。默认为 True。
- `title`：图表的标题。
- `ylabel`：y 轴的标签。

````python
# 准备数据,计算各类广告投入占比
data =  pd.read_csv('./data/advertising.csv')
pieData = data[['TV','Weibo','WeChat']]
sumData = pieData.sum()
sumData.plot(kind='pie',figsize=(6,6),title='Advertising Expenditure',fontsize=14,legend=False,autopct='%1.1f%%'
             , explode=[0,0.2,0],shadow=True,startangle=60)
plt.savefig('./out/output13.png',dpi=600)
````

   在绘制交互式地图之前，需要导入相关的模块。通常，我们需要导入 `Map` 类和 `Geo` 类来绘制地图。

   ```python
   from pyecharts.charts import Map, Geo
   from pyecharts import options as opts
   ```

3. 创建地图对象：
   使用 `Map` 类或 `Geo` 类创建一个地图对象。这些类提供了各种方法和选项，用于设置地图的样式、数据和交互功能。

   ```python
   # 使用 Map 类创建地图对象
   map_chart = Map()

   # 使用 Geo 类创建地图对象（适用于绘制全球地图）
   geo_chart = Geo()
   ```

4. 设置地图数据：
   通过 `add()` 方法向地图对象添加数据。可以传递地理位置名称和对应的数值。可以使用链式调用来添加多个数据点。

   ```python
   # 设置地图数据（示例）
   map_chart.add("地图标题", data_pair)
   ```

5. 设置地图样式和交互功能：
   使用 `set_global_opts()` 方法设置地图的全局样式和交互功能，例如标题、图例、视觉映射等。

   ```python
   # 设置地图全局样式和交互功能
   map_chart.set_global_opts(
       title_opts=opts.TitleOpts(title="地图标题"),
       visualmap_opts=opts.VisualMapOpts(),
       legend_opts=opts.LegendOpts(is_show=False)
   )
   ```

6. 渲染和保存地图：
   使用 `render()` 方法将地图渲染为 HTML 文件，或使用 `render_notebook()` 方法在 Jupyter Notebook 中渲染地图。你也可以使用 `save()` 方法将地图保存为图片文件。

```python
# 渲染地图为 HTML 文件
map_chart.render("map.html")
# 在 Jupyter Notebook 中渲染地图
map_chart.render_notebook()

# 将地图保存为图片文件
map_chart.render("map.png")
```

````python
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType

# 定义节点数据
lineFreq = [("上海", 723), ("广州", 43), ("北京", 128), ("乌鲁木齐", 12), ("重庆", 42)]
lineState = (
    Geo({"width": "1920px",
         "height": "1080px",
         })
    .add_schema(maptype="china")  #基本地图范围
    #添加显示城市序列（坐标点，值），显示颜色
    .add(
        "",
        lineFreq,
        type_=ChartType.EFFECT_SCATTER,
        color="#7EC0EE",
    )
    #添加箭头线序列（坐标点，坐标点），定义显得颜色和大小，角度
    .add(
        "",
        [("上海", "广州"), ("上海", "北京"), ("上海", "乌鲁木齐"), ("上海", "重庆")],
        type_=ChartType.LINES,
        effect_opts=opts.EffectOpts(
            symbol=SymbolType.ARROW, symbol_size=6, color="blue"
        ),
        linestyle_opts=opts.LineStyleOpts(curve=0.2),
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(title_opts=opts.TitleOpts(title="上海浦东机场至各地航班线路图"))
)
lineState.render("./out/geoline.html")
````
