```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import model_selection
from sklearn import preprocessing
```

# 特征值处理数据

```python
print(np.unique(df['Drug']) )
print(np.unique(df['Sex']) )
print(np.unique(df['Ascites']) )
print(np.unique(df['Hepatomegaly']) )
print(np.unique(df['Spiders']) )
print(np.unique(df['Edema']) )
```

```python
print(df.isnull().sum()) #均没有缺失值
```

1. **id**: 患者的唯一标识符。
2. **N_Days**: 患者在观察期间的存活时间（以天为单位）。
3. **Drug**: 患者接受的药物。
4. **Age**: 患者的年龄。
5. **Sex**: 患者的性别。
6. **Ascites**: 是否有腹水（是/否）。
7. **Hepatomegaly**: 是否有肝大（是/否）。
8. **Spiders**: 是否有蜘蛛痣（是/否）。
9. **Edema**: 是否有水肿（是/否）。N（无水肿且无利尿剂治疗）、S（有水肿未用利尿剂或已被利尿剂解决）、Y（尽管有利尿剂治疗仍有水肿）
10. **Bilirubin**: 胆红素水平。
11. **Cholesterol**: 胆固醇水平。
12. **Albumin**: 白蛋白水平。
13. **Copper**: 铜水平。
14. **Alk_Phos**: 碱性磷酸酶水平。
15. **SGOT**: 丙氨酸氨基转移酶水平。
16. **Tryglicerides**: 甘油三酯水平。
17. **Platelets**: 血小板计数。
18. **Prothrombin**: 凝血酶原时间。
19. **Stage**: 疾病的阶段。
20. **Status**: 患者的生存状态，包括C（被审查，表示患者在N_Days时仍然存活）、CL（表示患者在N_Days时因为进行了肝移植而仍然存活）和D（表示患者在N_Days时已经死亡）。

```python
# OBJECT->FLOAT
df.loc[df['Drug'] == 'D-penicillamine', 'Drug'] = 1 #青霉胺明显更严重
df.loc[df['Drug'] == 'Placebo', 'Drug'] = 0
df.loc[df['Sex'] == 'F', 'Sex'] = 1
df.loc[df['Sex'] == 'M', 'Sex'] = 0
df.loc[df['Ascites'] == 'N', 'Ascites'] = 0
df.loc[df['Ascites'] == 'Y', 'Ascites'] = 1
df.loc[df['Hepatomegaly'] == 'N', 'Hepatomegaly'] = 0
df.loc[df['Hepatomegaly'] == 'Y', 'Hepatomegaly'] = 1
df.loc[df['Spiders'] == 'N', 'Spiders'] = 0
df.loc[df['Spiders'] == 'Y', 'Spiders'] = 1
df.loc[df['Edema'] == 'N', 'Edema'] = 0
df.loc[df['Edema'] == 'S', 'Edema'] = 1
df.loc[df['Edema'] == 'Y', 'Edema'] = 2
```

Ascites,Hepatomegaly,Spiders,Edema特征合并

```python
df['sign_merged'] = df['Ascites'] + df['Spiders'] + df['Edema']+df['Hepatomegaly']+df['Drug']
df
```

```python
df.drop(['Ascites','Spiders','Edema','Hepatomegaly','Drug'], axis=1, inplace=True)
df
```

```python
# 对AGE和N-days 进行处理
df['Age'] = df['Age']/365
df['N_Days'] = df['N_Days']/365
df
```

```python
# 处理一下Status
# print(np.unique(df['Status']))
df.loc[df['Status']=='C','Status'] = 0
df.loc[df['Status']=='CL','Status']= 1 
df.loc[df['Status']=='D','Status'] =  2
df
```

```python
#进行 Max-min标准化
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
columns_to_fit = ['N_Days', 'Age', 'Bilirubin', 'Cholesterol', 'Albumin', 'Copper',
       'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets', 'Prothrombin',]
df[columns_to_fit] = scaler.fit_transform(df[columns_to_fit])
df
```

# 相关性分析

```python
corr = df.corr()['Status'].sort_values(ascending=False)
print(corr)
```

```python
#丢弃不重要数据
X = df.drop(['Status','id','Sex'], axis=1)

Y = df['Status'].astype(int)
```

# 异常值剔除

```python
from pyod.models.sod import SOD
sod = SOD(n_neighbors=20,ref_set=10,alpha=0.85,contamination=0.1)
sod.fit(X)
sod_lab = sod.labels_
sod_lab 
```

# 开始预测

```python
#切分数据集
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.2,shuffle=True,random_state=42)
```

```python
# 寻找max_depth
from sklearn.ensemble import RandomForestClassifier
d_scores = []
for i in range(1,20):
    RF = RandomForestClassifier(n_estimators=15,max_depth=i,random_state=42,)
    RF.fit(X_train,Y_train)
    d_scores.append(RF.score(X_test,Y_test))
depth = d_scores.index(max(d_scores))
print('决策树深度: ',depth,'最优值为: ',max(d_scores)) # 14
```

```python
n_scores = []
for i in range (1,105):
     RF = RandomForestClassifier(n_estimators=i,max_depth=11,random_state=42)
     RF.fit(X_train,Y_train)
     n_scores.append(RF.score(X_test,Y_test))
n_tree = n_scores.index(max(n_scores))
print('最优决策树数目: ',n_tree,'最优值为: ',max(n_scores))# 68
```

```python
from sklearn import metrics
#建立随机森林模型
rf = RandomForestClassifier(n_estimators=68,max_depth=11, random_state=42)
#模型训练
rf.fit(X_train, Y_train)
#进行预测
y_pred_rf = rf.predict(X_test)
#报告
class_report_rf = metrics.classification_report(y_true=Y_test, y_pred=y_pred_rf)
#评分:
score = rf.score(X_test,Y_test)
print("模型分数:",score)
print("模型报告")
print(class_report_rf)
```

# XGBOOST

```python
import xgboost as xgb
from xgboost.sklearn import XGBClassifier #sklearn xgboost
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_validate
X_train['sign_merged']= X_train['sign_merged'].astype(int)
X_test['sign_merged']= X_test['sign_merged'].astype(int)
```

```python


dTrain = xgb.DMatrix(X_train,label=Y_train)
dTest = xgb.DMatrix(X_test,label=Y_test)
params = {'eval_metric':'mlogloss','objective':'multi:softmax','eta':0.15,'max_depth':7,'subsample':0.8,'gamma':0.1,'num_class':3,'silent':True}
num_round = 180
bst = xgb.train(params,dTrain,num_round)
preds_ = bst.predict(dTest)
metrics.accuracy_score(Y_test,preds_)
```

## 参数调优

```python
param_test1 = {
    'max_depth':range(3,20,2),
 'min_child_weight':range(1,6,2)
}
gsearch1 = GridSearchCV(
   estimator=XGBClassifier(learning_rate = 0.1, n_estimators=140, max_depth=5,min_child_weight=1, gamma=0, subsample=0.8,
colsample_bytree=0.8,objective= 'multi:softmax', nthread=4,scale_pos_weight=1, seed=27,eval_metric='mlogloss'),
    param_grid=param_test1,n_jobs=4,cv=5
)
gsearch1.fit(X_train,Y_train)
gsearch1.cv_results_,gsearch1.best_params_,gsearch1.best_score_
```

```python
param_test3 = {
 'gamma':[i/10.0 for i in range(0,5)]
}
gsearch2 = GridSearchCV(
   estimator=XGBClassifier(learning_rate = 0.1, n_estimators=140, max_depth=7,min_child_weight=5, gamma=0, subsample=0.8,
colsample_bytree=0.8,objective= 'multi:softmax', nthread=4,scale_pos_weight=1, seed=27,eval_metric='mlogloss'),
    param_grid=param_test3,n_jobs=4,cv=5
)
gsearch2.fit(X_train,Y_train)
gsearch2.cv_results_,gsearch2.best_params_,gsearch2.best_score_
```

```python
param_test4 = {
 'subsample':[i/10.0 for i in range(6,20)],
 'colsample_bytree':[i/10.0 for i in range(6,20)]
}
gsearch3 = GridSearchCV(
   estimator=XGBClassifier(learning_rate = 0.1, n_estimators=140, max_depth=7,min_child_weight=5, gamma=0, subsample=0.8,
colsample_bytree=0.8,objective= 'multi:softmax', nthread=4,scale_pos_weight=1,num_class=3, seed=27,eval_metric='mlogloss'),
    param_grid=param_test4,n_jobs=4,cv=5
)
gsearch3.fit(X_train,Y_train)
gsearch3.cv_results_,gsearch3.best_params_,gsearch3.best_score_
```

```python
final_classifier=XGBClassifier(learning_rate = 0.1, n_estimators=140, max_depth=7,min_child_weight=5, gamma=0, subsample=1.0,
colsample_bytree=0.6,objective= 'multi:softmax', nthread=4,num_class=3, seed=27,eval_metric='mlogloss')
final_classifier.fit(X_train,Y_train)
final_train_pred = final_classifier.predict(X_test)
metrics.accuracy_score(Y_test,final_train_pred)
```
