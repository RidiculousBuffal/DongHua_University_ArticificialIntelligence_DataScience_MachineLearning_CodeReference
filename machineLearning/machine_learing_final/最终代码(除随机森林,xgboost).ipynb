# 头文件,读取数据,独热编码

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import model_selection
from sklearn import preprocessing
```

```python
df = pd.read_csv('train.csv')
print( df.dtypes) 
 # 需要进行独热编码的:
print(np.unique(df['Drug']) )
print(np.unique(df['Sex']) )
print(np.unique(df['Ascites']) )
print(np.unique(df['Hepatomegaly']) )
print(np.unique(df['Spiders']) )
print(np.unique(df['Edema']) )
# Drug,object
# Sex,object
# Ascites,object
# Hepatomegaly,object
# Spiders,object
# Edema,object
```

```python
df.loc[df['Drug']=='D-penicillamine','Drug'] = 0 
df.loc[df['Drug']=='Placebo','Drug']= 1 
df.loc[df['Sex']=='F','Sex'] =  1
df.loc[df['Sex']=='M','Sex'] =  0
df.loc[df['Ascites']=='N','Ascites'] = 0
df.loc[df['Ascites']=='Y','Ascites'] = 1
df.loc[df['Hepatomegaly']=='N','Hepatomegaly'] = 0
df.loc[df['Hepatomegaly']=='Y','Hepatomegaly'] = 1
df.loc[df['Spiders']=='N','Spiders'] = 0
df.loc[df['Spiders']=='Y','Spiders'] = 1
df.loc[df['Edema']=='N','Edema'] = 0
df.loc[df['Edema']=='S','Edema'] = -1
df.loc[df['Edema']=='Y','Edema'] = 1
```

```python
# dummies = pd.get_dummies(df[['Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema']],prefix=['Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema']).astype(int)
# df = df.drop(labels=['Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema'],axis=1).join(dummies)
# sum_na = df.isna().sum() #没有缺失值
```

```python
#把status转变为数值
print(np.unique(df['Status']))
df.loc[df['Status']=='C','Status'] = -1 
df.loc[df['Status']=='CL','Status']= 0 
df.loc[df['Status']=='D','Status'] =  1
```

#数据预处理,标准化 

```python
from sklearn.preprocessing import MinMaxScaler
print(df.columns)
# 需要标准化的列:
# 'N_Days', 'Age', 'Bilirubin', 'Cholesterol', 'Albumin', 'Copper',
#        'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets', 'Prothrombin',
#        'Stage' 
scaler = MinMaxScaler()
columns_to_fit = ['N_Days', 'Age', 'Bilirubin', 'Cholesterol', 'Albumin', 'Copper',
       'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets', 'Prothrombin',
       'Stage']
df[columns_to_fit] = scaler.fit_transform(df[columns_to_fit])
df
```

#观察相关系数,准备预测

```python
corr = df.corr()['Status'].sort_values(ascending=False)
print(corr)
# 性别,id Drug 与status关系不大
# df = df.drop(labels = ['id','Sex_F','Sex_M','Drug_D-penicillamine','Drug_Placebo'],axis=1)
df = df.drop(labels = ['id'],axis=1)
```

# 开始预测

```python
X = df.drop('Status',axis=1)
Y = df['Status'].astype(int)
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.5,random_state=20)
```

## 逻辑回归

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

logreg = LogisticRegression(max_iter=10000, random_state=42)
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
print("模型分数:",logreg.score(X_test, Y_test))
class_report = metrics.classification_report(y_pred=Y_pred,y_true=Y_test,zero_division=0)
print("报告")
print(class_report)
```

## SVM支持向量机
### 线性模型

```python
from sklearn.svm import SVC
svm_model = SVC(kernel='linear')

# 在训练集上训练模型
svm_model.fit(X_train, Y_train)
# 在测试集上进行预测
y_pred_svm_linear = svm_model.predict(X_test)
print("模型分数:",svm_model.score(X_test, Y_test))
print("模型报告:",metrics.classification_report(y_pred=y_pred_svm_linear,y_true=Y_test,zero_division=0))
```

## 多项式模型


```python
svm_model_poly = SVC(kernel='poly',degree=3,coef0=1)


# 在训练集上训练模型
svm_model_poly.fit(X_train, Y_train)
y_pred_svm_ploy = svm_model_poly.predict(X_test)
print("模型分数:",svm_model_poly.score(X_test, Y_test))
print("模型报告:",metrics.classification_report(y_pred=y_pred_svm_ploy,y_true=Y_test,zero_division=0))
```

### 径向基函数核函数

```python
svm_model_rbf = SVC(kernel='rbf', gamma=0.3)
# 在训练集上训练模型
svm_model_rbf.fit(X_train, Y_train)
y_pred_svm_rbf = svm_model_rbf.predict(X_test)
print("模型分数:",svm_model_rbf.score(X_test, Y_test))
print("模型报告:",metrics.classification_report(y_pred=y_pred_svm_rbf,y_true=Y_test,zero_division=0))
```

## 随机森林模型

```python
# 寻找max_depth
from sklearn.ensemble import RandomForestClassifier
d_scores = []
for i in range(1,20):
    RF = RandomForestClassifier(n_estimators=15,max_depth=i,random_state=42,criterion='entropy')
    RF.fit(X_train,Y_train)
    d_scores.append(RF.score(X_test,Y_test))
depth = d_scores.index(max(d_scores))
print('决策树深度: ',depth,'最优值为: ',max(d_scores)) # 11
```

```python
# 按照最优深度,找最优决策树数目
n_scores = []
for i in range (1,105):
     RF = RandomForestClassifier(n_estimators=i,max_depth=11,random_state=42,criterion='entropy')
     RF.fit(X_train,Y_train)
     n_scores.append(RF.score(X_test,Y_test))
n_tree = n_scores.index(max(n_scores))
print('最优决策树数目: ',n_tree,'最优值为: ',max(n_scores))# 75
```

```python
#建立随机森林模型
rf = RandomForestClassifier(n_estimators=82,max_depth=10, random_state=42)
#模型训练
rf.fit(X_train, Y_train)
#进行预测
y_pred_rf = rf.predict(X_test)
#报告
class_report_rf = metrics.classification_report(y_true=Y_test, y_pred=y_pred_rf)
#评分:
score = rf.score(X_test,Y_test)
print("模型分数:",score)
print("模型报告")
print(class_report_rf)
```

```python
n_scores = []
for i in range (1,105):
     RF_pca = RandomForestClassifier(n_estimators=i,max_depth=8,random_state=42,criterion='entropy')
     RF_pca.fit(X_train_pca,Y_train)
     n_scores.append(RF_pca.score(X_test_pca,Y_test))
n_tree = n_scores.index(max(n_scores))
print('最优决策树数目: ',n_tree,'最优值为: ',max(n_scores))# 39
```
